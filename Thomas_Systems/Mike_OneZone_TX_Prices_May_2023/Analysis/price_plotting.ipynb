```python
import pandas as pd
import matplotlib.pyplot as plt

# Read in the first CSV file
df1 = pd.read_csv('2019_empirical_electricity_prices.csv')

# Read in the second CSV file
df2 = pd.read_csv('prices.csv')

# Add a Time_Index column to df2
df2['Time_Index'] = range(min(df1['Time_Index']), max(df1['Time_Index'])+1)

# Plot both data frames on the same graph
plt.plot(df1['Time_Index'], df1['Prices'])
plt.plot(df2['Time_Index'], df2['1'])
plt.xlabel('Hours')
plt.ylabel('Prices')
plt.legend(['Empirical', 'GenX'])
plt.show()
```
